```python
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib.image import imread
from itertools import chain
import numpy as np
import scipy as sp
import glob
```

```python
filenames = [img for img in glob.glob("fotos_eigenfaces*/*.pgm")]
m = [[] for i in range(16)]
for i in range(16):
  m[i] = list(chain.from_iterable(imread(filenames[i])))
  
m = np.matrix(m)      
```

```python
model = PCA(n_components=14)
pts = normalize(m)
model.fit(pts)
pts2 = model.transform(pts)
```

```python
fig, ax = plt.subplots(1,14)
for i in range(14):
  ord = model.components_[i]
  img = ord.reshape(64,64)
  ax[i].imshow(img,cmap='gray')
```

```python
faces_pca = PCA(n_components=8)
faces_pca.fit(m)
    
components = faces_pca.transform(m)
projected = faces_pca.inverse_transform(components)
fig, axes = plt.subplots(4,4,figsize=(20,20))
for i, ax in enumerate(axes.flat):
    ax.imshow(projected[i].reshape(64,64),cmap="gray")
```
