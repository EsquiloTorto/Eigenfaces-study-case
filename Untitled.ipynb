<h1 style="text-align: center">EIGENFACES</h1>
<hr>
    
    Um dos usos mais icônicos da decomposição SVD de matrizes é chamado de eigenfaces. É utilizado a decomposição PCA (uma versão reduzida do SVD, a qual é considerado apenas os elementos mais significativos para a matriz final ) em um dataset de treinamento que contém imagens variadas de pessoas. Resultando em uma quantidade x de eigenfaces, que servirão como base para reconstruir imagens.

<h3>importando bibliotecas</h3>

```python
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib.image import imread
from itertools import chain
import numpy as np
import scipy as sp
import glob
```

<h3>Construindo a matriz de imagens</h3>

Cada imagem possui uma largura de m pixels e altura de n pixels (m x n). Para facilitar o uso das imagens em uma matriz, cada uma será transformada em um vetor coluna (1 x mn). Como o ilustrado na imagem:
<div style="text-align: center">
    <img src="imagens/imagem1.png" alt="ilustração das imagens em vetores" width=400>
</div>

```python
# Cria uma lista com todos os nomes das imagens
filenames = [img for img in glob.glob("fotos_eigenfaces*/*.pgm")]

# Cria uma lista vazia com a quantidade de imagens na pasta
m = [[] for i in range(len(filenames))]

# Transforma as imagens 64x64 em vetores de 4096 linhas e transforma em uma matriz numpy
for i in range(16):
  m[i] = list(chain.from_iterable(imread(filenames[i])))
m = np.matrix(m)      
```

Temos 16 imagens e sendo assim nossa matriz terá 16 colunas. A quantidade de linhas é definida pela multiplicação da largura pela altura da imagem em pixels. Nesse caso 64 x 64 = 4096

```python
formato = m.shape
print(formato)
```

<h3>Criando os eigenfaces</h3>

```python
model = PCA(n_components=12)
pts = normalize(m)
model.fit(pts)
pts2 = model.transform(pts)
```

```python
fig, ax = plt.subplots(3,4,figsize=[32,32])
for i in range(12):
  ord = model.components_[i]
  img = ord.reshape(64,64)
  if i < 4:
    ax[0,i].imshow(img,cmap='gray')
  elif i < 8:
    ax[1,i-4].imshow(img,cmap='gray')
  elif i < 12:
    ax[2,i-8].imshow(img,cmap='gray')
```

<h3>reconstruindo as imagens do banco</h3>

```python
faces_pca = PCA(n_components=12)
faces_pca.fit(m)
    
components = faces_pca.transform(m)
projected = faces_pca.inverse_transform(components)
fig, axes = plt.subplots(4,4,figsize=(20,20))
for i, ax in enumerate(axes.flat):
    ax.imshow(projected[i].reshape(64,64),cmap="gray")
```
